```python
#question 2 
#showing both separately 
from scipy.stats import norm
from scipy.special import comb
import matplotlib.pyplot as plt
import numpy as np
import math
import textwrap
n_experiments = 100000
a_values = np.random.uniform(0, 100, size=4)
p = 0.2
for a in a_values:
    n = int(np.ceil(a / p))
    poisson = np.random.poisson(a, n_experiments)
    binomial = np.random.binomial(n, p, n_experiments)
    fig, ax = plt.subplots()
    ax.hist(poisson, bins=np.arange(0, max(poisson) + 1), alpha=0.5, label='Poisson')
    ax.legend()
    ax.set_xlabel('poisson experiment outcome')
    ax.set_ylabel('Frequency of outcomes')
    ax.set_title(f'aplha={a}')
    plt.show()
    fig, ax = plt.subplots()
    ax.hist(binomial, bins=np.arange(0, max(binomial) + 1), alpha=0.5, label='Binomial')
    ax.legend()
    ax.set_xlabel('binomial experiment outcome')
    ax.set_ylabel('Frequency of outcomes')
    ax.set_title(f'number trials={n}')
    plt.show()
```

```python
# showing the overlapping of both the  RVs 
from scipy.stats import norm
from scipy.special import comb
import matplotlib.pyplot as plt
import numpy as np
import math
import textwrap

n_experiments = 100000
a_values = np.random.uniform(0, 100, size=4)
p = 0.2

for a in a_values:
    n = int(np.ceil(a / p))
    poisson = np.random.poisson(a, n_experiments)
    binomial = np.random.binomial(n, p, n_experiments)
    fig, ax = plt.subplots()
    ax.hist(poisson, bins=np.arange(0, max(poisson) + 1), alpha=0.5, label='Poisson')
    ax.hist(binomial, bins=np.arange(0, max(binomial) + 1), alpha=0.5, label='Binomial')
    
    ax.legend()
    ax.set_xlabel('binomial and poisson experiment outcome')
    ax.set_ylabel('Frequency of outcomes')
    plt.title(f'a={a}, n={n}')
    plt.show()
```

```python
# Getting Python ready
from scipy.stats import norm
from scipy.special import comb
import matplotlib.pyplot as plt
import numpy as np
import math
import textwrap


lambda_val = np.random.randint(1, 21, size=2)
seq_num = [100, 500, 1000, 10000]

def poisson_sum(length, lambdas):
    s_means = []
    for i in range(length):
        poisson_vals = np.random.poisson(lam=lambdas, size=length)
        s_sum = np.sum(poisson_vals)
        s_mean = s_sum / len(poisson_vals)
        s_means.append(s_mean)
    return s_means

fig, axs = plt.subplots(len(seq_num), len(lambda_val), figsize=(12, 8))
for i, length in enumerate(seq_num):
    for j, lambd in enumerate(lambda_val):
        s_means = poisson_sum(length, lambd)
        axs[i, j].hist(s_means, bins=30, density=True)
        axs[i, j].set_title(f"n = {length},alpha = {lambd}")
        axs[i, j].set_xlabel("no of arrivals")
        axs[i, j].set_ylabel("Frequency of outcomes")
        mean_value = np.mean(s_means)
        std_value = np.std(s_means)
        x = np.linspace(mean_value - 3 * std_value, mean_value + 3 * std_value, 100)
        y = norm.pdf(x, mean_value, std_value)
        axs[i, j].plot(x, y, color='red')  

plt.tight_layout()
plt.show()
#observation
#in this as  n is increasing, the curve is resembling the bell shaped curve and becoming more narrower.    
```
